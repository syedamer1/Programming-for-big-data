# Q1) Find the total number of patients of each disease (diagnosis) and plot in histogram.

```python
import pandas as pd
df = pd.read_csv('jsrt_metadata.csv')
print(df['diagnosis'].value_counts())
print("Total Number of Patients:",df['diagnosis'].value_counts().sum())
```

```python
import matplotlib as plt
from matplotlib import pyplot

df['diagnosis']=df['diagnosis'].astype('str')
plt.pyplot.hist(df['diagnosis'],color='blue',ec='pink',orientation='horizontal')
plt.pyplot.title('Number of People Diagnosis')
plt.pyplot.xlabel('Frequency')
plt.pyplot.ylabel('Disease Category')
```

# Q2) . Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.

```python
print(df[['diagnosis','state']].value_counts())
```

```python
plt.pyplot.pie(df[['diagnosis','state']].value_counts(),shadow=True,autopct='%1.1f%%')
```
