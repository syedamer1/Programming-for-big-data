```python
import pandas as pd
```

```python
df = pd.read_csv('file.csv')
data = df['Data'].tolist()
size = len(df['Data'])
```

```python
def calculatedistance(x1, x2, datalist, si):
    c_1 = [] 
    c_2 = [] 
    for x in range(0, si):
        c_1.append(abs(int(datalist[x] - x1)))
        c_2.append(abs(int(datalist[x] - x2)))
    return (c_1, c_2)

def checkc1c2(x, y):
    if x < y:
        return 1
    elif y < x:
        return 2
    elif x == y:
        return 3
```

```python
import random
centralpoint1 = []
centralpoint2 = []
s = 'A'
precheck = []
currcheck = []
check = False
c1_ = []
c2_ = []
c1 = random.choice(data)
centralpoint1.append(c1)
c2 = random.choice(data)
centralpoint2.append(c2)
(c1_, c2_) = calculatedistance(c1, c2, data, size)
for i in range(0, size):
    result = checkc1c2(c1_[i], c2_[i])
    if result == 1:
        currcheck.append('c1less')
    elif result == 2:
        currcheck.append('c2less')
    elif result == 3:
        currcheck.append('equ')
precheck = currcheck
df[s + 'C' + str(1)] = c1_
df[s + 'C' + str(2)] = c2_ 
s = chr(ord(s) + 1)
while True:
    c1_ = []
    c2_ = []
    c1 = random.choice(data)
    centralpoint1.append(c1)
    c2 = random.choice(data)
    centralpoint2.append(c2)
    (c1_, c2_) = calculatedistance(c1, c2, data, size)
    currcheck = []
    for i in range(0, size):
        result = checkc1c2(c1_[i], c2_[i])
        if result == 1:
            currcheck.append('c1less')
        elif result == 2:
            currcheck.append('c2less')
        elif result == 3:
            currcheck.append('equ')
    check = False
    for j in range(0, size):
        if precheck[j] == 'c2less' and currcheck[j] == 'c1less':
            check = False
            break
        elif precheck[j] == 'c1less' and currcheck[j] == 'c2less':
            check = False
            break
        elif precheck[j] == 'equ' and currcheck[j] == 'c1less':
            check = True
        elif precheck[j] == 'equ' and currcheck[j] == 'c2less':
            check = True
        elif precheck[j] == 'equ' and currcheck[j] == 'equ':
            check = True
        elif precheck[j] == 'c1less' and currcheck[j] == 'equ':
            check = True
        elif precheck[j] == 'c2less' and currcheck[j] == 'equ':
            check = True
        elif precheck[j] == currcheck[j]:
            check = True
    df[s + 'C' + str(1)] = c1_
    df[s + 'C' + str(2)] = c2_ 
    s = chr(ord(s) + 1)
    if s == '[':
        s = 'A'
    if check == True:
        break
    precheck = currcheck
```

```python
print("Central Points of C1: ",centralpoint1)
print("Central Points of C2: ",centralpoint2)
```
